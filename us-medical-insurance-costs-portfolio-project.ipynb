# U.S. Medical Insurance Costs

# Scope
We are trying to investigate whether region affects insurance costs and if yes, why that is so. 

## Investigation
First, we can find the average insurance cost for the different regions to see if there is indeed a difference in insurance cost. 


```python

import csv
```

```python
#Converting csv to a list of dictionaries which store the individuals' information
with open("C:\\Users\\cheng\\AppData\\Local\\Temp\\bfad068c-1cda-40ae-bec1-80960efb058b_python-portfolio-project-medical-insurance-cost-starter-files.zip.58b\\python-portfolio-project-starter-files\\insurance.csv") as csvfile:
    csvreader = csv.DictReader(csvfile)
    data = [row for row in csvreader]
    
print(data)
    
```

```python
#find the distinct regions
region_list = [person['region'] for person in data]
def distinct(list1):
    unique_region_list = []
    for region in list1:
        if region not in unique_region_list:
            unique_region_list.append(region)
    for region in unique_region_list:
        print(region)

distinct(region_list)

```

```python
#find population sorted by regions
southwest_population = 0
southeast_population = 0
northwest_population = 0
northeast_population = 0

for person in data:
    if person['region'] == 'northeast':
        northeast_population += 1
    if person['region'] == 'northwest':
        northwest_population += 1
    if person['region'] == 'southwest':
        southwest_population += 1
    if person['region'] == 'southeast':
        southeast_population += 1

print('southwest population is ' + str(southwest_population))
print('northwest population is ' + str(northwest_population))
print('northeast population is ' + str(northeast_population))
print('southeast population is ' + str(southeast_population))
```

```python
# average insurance cost sorted by region
southwest_cost = 0
southeast_cost = 0
northwest_cost = 0
northeast_cost = 0

for person in data:
    if person['region'] == 'northeast':
        northeast_cost += float(person['charges'])
    if person['region'] == 'northwest':
        northwest_cost += float(person['charges'])
    if person['region'] == 'southwest':
        southwest_cost += float(person['charges'])
    if person['region'] == 'southeast':
        southeast_cost += float(person['charges'])

print('southwest average insurance cost is ' + str(southwest_cost/southwest_population))
print('northwest average insurance cost is ' + str(northwest_cost/northwest_population))
print('northeast average insurance cost is ' + str(northeast_cost/northeast_population))
print('southeast average insurance cost is ' + str(southeast_cost/southeast_population))
```

# Results 
The people of southeast are on average paying up to 2500 dollars more than their counterparts in other regions. 
This could be due to multiple factors such as 
    1. percentage of affluent residents in this region resulting in insurance companies marking up their insurance plans
    2. there could be more people that are classified as higher risk individuals (smokers, higher bmi etc)

So we have to investigate what are the factors that result in this increase in insurance charges.

```python
#number of smokers sorted by region

southwest_smokers = 0
southeast_smokers = 0
northwest_smokers = 0
northeast_smokers = 0

for person in data:
    if person['region'] == 'northeast' and person['smoker'] == 'yes':
        northeast_smokers += 1
    if person['region'] == 'southeast' and person['smoker'] == 'yes':
        southeast_smokers += 1
    if person['region'] == 'northwest' and person['smoker'] == 'yes':
        northwest_smokers += 1
    if person['region'] == 'southwest' and person['smoker'] == 'yes':
        southwest_smokers += 1

print('southwest has ' + str(southwest_smokers) + ' smokers')
print('northwest has ' + str(northwest_smokers) + ' smokers')
print('northeast has ' + str(northeast_smokers) + ' smokers')
print('southeast has ' + str(southeast_smokers) + ' smokers')
```

```python
#average bmi sorted by regions
southwest_bmis = 0
southeast_bmis = 0
northwest_bmis = 0
northeast_bmis = 0

for person in data:
    if person['region'] == 'northeast':
        northeast_bmis += float(person['bmi'])
    if person['region'] == 'northwest':
        northwest_bmis += float(person['bmi'])
    if person['region'] == 'southwest':
        southwest_bmis += float(person['bmi'])
    if person['region'] == 'southeast':
        southeast_bmis += float(person['bmi'])

print('southwest average bmi is ' + str(southwest_bmis/southwest_population))
print('northwest average bmi is ' + str(northwest_bmis/northwest_population))
print('northeast average bmi is ' + str(northeast_bmis/northeast_population))
print('southeast average bmi is ' + str(southeast_bmis/southeast_population))
```

```python
#average number of children sorted by region
southwest_kids = 0
southeast_kids = 0
northwest_kids = 0
northeast_kids = 0

for person in data:
    if person['region'] == 'northeast':
        northeast_kids += float(person['children'])
    if person['region'] == 'northwest':
        northwest_kids += float(person['children'])
    if person['region'] == 'southwest':
        southwest_kids += float(person['children'])
    if person['region'] == 'southeast':
        southeast_kids += float(person['children'])

print('southwest average no of children is ' + str(southwest_kids/southwest_population))
print('northwest average no of children is ' + str(northwest_kids/northwest_population))
print('northeast average no of children is ' + str(northeast_kids/northeast_population))
print('southeast average no of children is ' + str(southeast_kids/southeast_population))
```

```python
#average age sorted by region
southwest_ages = 0
southeast_ages = 0
northwest_ages = 0
northeast_ages = 0

for person in data:
    if person['region'] == 'northeast':
        northeast_ages += float(person['age'])
    if person['region'] == 'northwest':
        northwest_ages += float(person['age'])
    if person['region'] == 'southwest':
        southwest_ages += float(person['age'])
    if person['region'] == 'southeast':
        southeast_ages += float(person['age'])

print('southwest average age is ' + str(southwest_ages/southwest_population))
print('northwest average age is ' + str(northwest_ages/northwest_population))
print('northeast average age is ' + str(northeast_ages/northeast_population))
print('southeast average age is ' + str(southeast_ages/southeast_population))
```

```python
#ratio of genders sorted by regions
southwest_males = 0
southeast_males = 0
northwest_males = 0
northeast_males = 0

for person in data:
    if person['region'] == 'northeast' and person['sex'] == 'male':
        northeast_males += 1
    if person['region'] == 'northwest' and person['sex'] == 'male':
        northwest_males += 1
    if person['region'] == 'southwest' and person['sex'] == 'male':
        southwest_males += 1
    if person['region'] == 'southeast' and person['sex'] == 'male':
        southeast_males += 1

print('southwest percentage of males is ' + str((southwest_males/southwest_population) * 100) + '%')
print('northwest percentage of males is ' + str((northwest_males/northwest_population) * 100) + '%')
print('northeast percentage of males is ' + str((northeast_males/northeast_population) * 100) + '%')
print('southeast percentage of males is ' + str((southeast_males/southeast_population) * 100) + '%')
```

```python
import matplotlib.pyplot as plt
fig, (age_ax, sex_ax, bmi_ax, children_ax, smoker_ax, charges_ax) = plt.subplots(6, 1, sharex=True)

age_ax.plot(['southwest', 'northwest', 'northeast', 'southeast'], [southwest_ages/southwest_population, northwest_ages/northwest_population, northeast_ages/northeast_population, southeast_ages/southeast_population])
age_ax.set_ylabel('Average Age')

sex_ax.plot(['southwest', 'northwest', 'northeast', 'southeast'], [(southwest_males/southwest_population) * 100, (northwest_males/northwest_population) * 100, (northeast_males/northeast_population) * 100, (southeast_males/southeast_population) * 100])
sex_ax.set_ylabel('Percentage of males')

bmi_ax.plot(['southwest', 'northwest', 'northeast', 'southeast'], [southwest_bmis/southwest_population, northwest_bmis/northwest_population, northeast_bmis/northeast_population, southeast_bmis/southeast_population])
bmi_ax.set_ylabel('Average bmi')

children_ax.plot(['southwest', 'northwest', 'northeast', 'southeast'], [southwest_kids/southwest_population, northwest_kids/northwest_population, northeast_kids/northeast_population, southeast_kids/southeast_population])
children_ax.set_ylabel('Average no of children')

charges_ax.plot(['southwest', 'northwest', 'northeast', 'southeast'], [southwest_cost/southwest_population, northwest_cost/northwest_population, northeast_cost/northeast_population, southeast_cost/southeast_population])
charges_ax.set_ylabel('Charges')

smoker_ax.plot(['southwest', 'northwest', 'northeast', 'southeast'], [southwest_smokers, northwest_smokers, northeast_smokers, southeast_smokers])
smoker_ax.set_ylabel('No of smokers')
smoker_ax.set_xlabel('Regions')

plt.subplots_adjust(left=0.1, right=0.9, bottom=0, top=2, hspace=1,)
plt.show()





```
