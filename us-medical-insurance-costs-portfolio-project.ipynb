# U.S. Medical Insurance Costs

# Scope
We are trying to investigate whether region affects insurance costs and if yes, why that is so. 

## Investigation
First, we can find the average insurance cost for the different regions to see if there is indeed a difference in insurance cost. 


```python

import csv
```

```python
#Converting csv to a list of dictionaries which store the individuals' information
with open("C:\\Users\\cheng\\AppData\\Local\\Temp\\58913cb5-4783-4cca-a099-026011b2ed67_python-portfolio-project-medical-insurance-cost-starter-files.zip.d67\\python-portfolio-project-starter-files\\insurance.csv") as csvfile:
    csvreader = csv.DictReader(csvfile)
    data = [row for row in csvreader]
    
print(data)
    
```

```python
#find the distinct regions
region_list = [person['region'] for person in data]
def distinct(list1):
    unique_region_list = []
    for region in list1:
        if region not in unique_region_list:
            unique_region_list.append(region)
    for region in unique_region_list:
        print(region)

distinct(region_list)

```

```python
#find population sorted by regions
southwest_population = 0
southeast_population = 0
northwest_population = 0
northeast_population = 0

for person in data:
    if person['region'] == 'northeast':
        northeast_population += 1
    if person['region'] == 'northwest':
        northwest_population += 1
    if person['region'] == 'southwest':
        southwest_population += 1
    if person['region'] == 'southeast':
        southeast_population += 1

print('southwest population is ' + str(southwest_population))
print('northwest population is ' + str(northwest_population))
print('northeast population is ' + str(northeast_population))
print('southeast population is ' + str(southeast_population))
```

```python
# average insurance cost sorted by region
southwest_cost = 0
southeast_cost = 0
northwest_cost = 0
northeast_cost = 0

for person in data:
    if person['region'] == 'northeast':
        northeast_cost += float(person['charges'])
    if person['region'] == 'northwest':
        northwest_cost += float(person['charges'])
    if person['region'] == 'southwest':
        southwest_cost += float(person['charges'])
    if person['region'] == 'southeast':
        southeast_cost += float(person['charges'])

print('southwest average insurance cost is ' + str(southwest_cost/southwest_population))
print('northwest average insurance cost is ' + str(northwest_cost/northwest_population))
print('northeast average insurance cost is ' + str(northeast_cost/northeast_population))
print('southeast average insurance cost is ' + str(southeast_cost/southeast_population))
```

# Results 
The people of southeast are on average paying up to 2500 dollars more than their counterparts in other regions. 
This could be due to multiple factors such as <br>
    1. percentage of affluent residents in this region resulting in insurance companies marking up their insurance plans<br>
    2. there could be more people that are classified as higher risk individuals (smokers, higher bmi etc)

So we have to investigate what are the factors that result in this increase in insurance charges.

```python
#number of smokers sorted by region

southwest_smokers = 0
southeast_smokers = 0
northwest_smokers = 0
northeast_smokers = 0

for person in data:
    if person['region'] == 'northeast' and person['smoker'] == 'yes':
        northeast_smokers += 1
    if person['region'] == 'southeast' and person['smoker'] == 'yes':
        southeast_smokers += 1
    if person['region'] == 'northwest' and person['smoker'] == 'yes':
        northwest_smokers += 1
    if person['region'] == 'southwest' and person['smoker'] == 'yes':
        southwest_smokers += 1                                          

print('southwest percentage of smokers is ' + str(float(southwest_smokers/southwest_population * 100)))
print('northwest percentage of smokers is ' + str(float(northwest_smokers/northwest_population * 100)))
print('northeast percentage of smokers is ' + str(float(northeast_smokers/northeast_population * 100)))
print('southeast percentage of smokers is ' + str(float(southeast_smokers/southwest_population * 100)))
```

```python
#average number of children sorted by region
southwest_kids = 0
southeast_kids = 0
northwest_kids = 0
northeast_kids = 0

for person in data:
    if person['region'] == 'northeast':
        northeast_kids += float(person['children'])
    if person['region'] == 'northwest':
        northwest_kids += float(person['children'])
    if person['region'] == 'southwest':
        southwest_kids += float(person['children'])
    if person['region'] == 'southeast':
        southeast_kids += float(person['children'])

print('southwest average no of children is ' + str(southwest_kids/southwest_population))
print('northwest average no of children is ' + str(northwest_kids/northwest_population))
print('northeast average no of children is ' + str(northeast_kids/northeast_population))
print('southeast average no of children is ' + str(southeast_kids/southeast_population))
```

```python
#ratio of genders sorted by regions
southwest_males = 0
southeast_males = 0
northwest_males = 0
northeast_males = 0

for person in data:
    if person['region'] == 'northeast' and person['sex'] == 'male':
        northeast_males += 1
    if person['region'] == 'northwest' and person['sex'] == 'male':
        northwest_males += 1
    if person['region'] == 'southwest' and person['sex'] == 'male':
        southwest_males += 1
    if person['region'] == 'southeast' and person['sex'] == 'male':
        southeast_males += 1

print('southwest percentage of males is ' + str((southwest_males/southwest_population) * 100) + '%')
print('northwest percentage of males is ' + str((northwest_males/northwest_population) * 100) + '%')
print('northeast percentage of males is ' + str((northeast_males/northeast_population) * 100) + '%')
print('southeast percentage of males is ' + str((southeast_males/southeast_population) * 100) + '%')
```

# Conclusion
- Generally, we can see that for % of smokers, average bmi and % of males, the graphs have a similar trend and shape which is that the higher the % of smokers, average bmi and % of males, the higher the insurance charges are. This goes as expected as these are the people that are of higher risk and hence are more likely to incur medical charges, hence their insurance premiums will increase accordingly.

- We also notice that this is not the case for average number of children and average age despite the fact that it should also have followed a similar trend. 
    - It could be that the differences across the regions are too insignificant compared to the differences in the previous three factors as seen in the distribution of the average age across the regions where the average ages lie within the same age of 39.
    - It could also be that the insurance companies do not factor in these factors too highly in their medical insurance charges formula hence not significantly counteracting the increase in insurance charges from the previous three factors which clearly shows positive correlation with the insurance charges.

- That being said, this is a very surface level analysis. To get a clearer picture, we could consider the distributions of the data and look at the mean, standard deviation etc to consider if the differences are possibly significant enough to have cause those differences in insurance charges. For example, although the average ages are around the same, it maybe the case where there might be more people in the extremely ends of the age spectrum such as youths and elderly. We roughly know that as age increases, the insurance charges increases as well. Depending on how the insurance companies calculate their insurance charges (ie whether it is disproportionately more costly for the elderly compared to how much cheaper it is for youths to get health insurance), this may affect the average insurance charges.

- More importantly, given the data, we do not know if region itself has an effect on insurance charges, whether it could be that insurance companies already have assumptions on people that live in a certain region. One assumption could be a certain region may be associated with higher percentage of affluent people which results in insurance companies marking up their insurance plans. So, we might need to consider whether income/net worth (basically any indication of wealth) affects insurance costs. Also, we are assuming that these are the only factors affecting the insurance charges, when in reality, insurance companies may consider alot more than just these few factors.
